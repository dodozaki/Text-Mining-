Il r√©cup√®re des √©l√©ments tels que les profils d'utilisateurs, les hashtags ou les recherches et renvoie les √©l√©ments d√©couverts.  On a collect√© 500 tweets porte le mot ‚Äòangoisse‚Äô pour la date 2022/02/05 et l‚Äôemport√© sous forme csv.


```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('anguish until:2022-02-05').get_items()):
    if i>500:
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount, tweet.user.displayname, tweet.lang])
    
# Creating a dataframe from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Like Count', 'Display Name', 'Language'])
```

```python
tweets_df1.to_csv('data extraction.csv')
```
